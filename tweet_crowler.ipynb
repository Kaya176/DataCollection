기존 트윗 데이터를 (무료로)수집하기 위해서는 상당한 제약이 존재합니다.  
1. 7일전 까지 발생한 트윗들만 수집 가능

이러한 문제를 해결하기 위해 github에는 상당히 많은 트윗 데이터 크롤링을 위한 소스코드가 있지만..  
아쉽게도 최근 twitter측에서 접근을 다 막은듯 하다...  

하지만, 다행이도 아직 작동하는 크롤링 tool이 남아 있었다!

Scweet(github) : https://github.com/Altimis/Scweet

위의 패키지를 이용하면 아무런 제약없이(참고: legal tool임) 트윗 데이터를 마음대로 수집할 수 있다.  

수집되는 데이터 정보는 다음과 같다.  

```
>> [UserScreenName, UserName, Timestamp, Text, Embedded_text, Emogis, Comments, Likes, Retweets,Image link, Tweet URL]
```
본격적으로 크롤링을 위해 설치해보자!

## [Scweet 설치 방법]

### [1] pip를 이용하여 설치하기(EASY)
```
>>pip install scweet 
```

### [2] github를 이용하여 설치하기
```
>>git clone https://github.com/Altimis/Scweet.git
>>cd ~/Scweet
>>python setup.py install
```
---
## How to Use ?

### [1] 특정 keyword를 이용하여 크롤링 하기  
```
from Scweet.scweet import scrap

data = scrap(words=['covid','covid19'], start_date="2020-04-01", max_date="2020-04-15", from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, 
             resume=False, filter_replies=True, proximity=True)
```
### [2] Hashtag를 이용하여 크롤링 하기
```
from Scweet.scweet import scrap

data = scrap(hashtag="covid19", start_date="2020-04-01", max_date="2020-04-15", from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, 
             resume=False, filter_replies=True, proximity=True)
```
